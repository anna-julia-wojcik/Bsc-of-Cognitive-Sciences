### $\color{#FF66B2}{\textbf{Task description}}$

**Goal:** Create a function `calculate(a, op, b)` that performs a mathematical operation on two numbers `a` and `b`, where `op` is one of the operators:
`+ - * / // % **`

- `a` and `b` are numbers (`int` or `float`).  
- `op` is a string (e.g., `"+"`).  

The function should handle **errors** (e.g., division by zero) and provide **clear messages**.

Think of it as a small â€œmachineâ€: you input `a`, `op`, `b` â†’ it outputs either the result or a friendly error message.

### $\color{#FF66B2}{\textbf{Constants and my own calculator error}}$

```python
# Set of allowed operators
ALLOWED_OPERATORS = {"+", "-", "*", "/", "//", "%", "**"}

# My own calculator error, so the user gets clear messages
class CalculatorError(Exception):
    """Calculator error â€” eg. not allowed operator or divison by zero."""
    pass
```

### $\color{#FF66B2}{\textbf{Main function - calculate}}$

```python
def calculate(a: int | float, op: str, b: int | float) -> int | float:
    """
    Return the result of the operation: a <op> b.

    Args:
        a, b: numbers (int or float)
        op: text, one of ALLOWED_OPERATORS
    Returns:
        int | float: result of the operation
    """
    # Validate types
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Arguments must be numbers (int or float).")

    # Validate operator
    if op not in ALLOWED_OPERATORS:
        raise CalculatorError(f"Unknown operator: {op}")

    # Check division by zero
    if op in ["/", "//", "%"] and b == 0:
        raise CalculatorError("Division by zero")

    # Perform operation using match-case (Python 3.10+)
    match op:
        case "+":
            return a + b
        case "-":
            return a - b
        case "*":
            return a * b
        case "/":
            return a / b
        case "//":
            return a // b
        case "%":
            return a % b
        case "**":
            return a ** b
        case _:
            raise CalculatorError("Unknown operator usage.")
```

### $\color{#FF66B2}{\textbf{Basic tests}}$

```python
assert calculate(2, "+", 3) == 5
assert calculate(2, "-", 5) == -3
assert calculate(3, "*", 4) == 12
assert calculate(7, "/", 2) == 3.5
assert calculate(7, "//", 2) == 3
assert calculate(7, "%", 3) == 1
assert calculate(2, "**", 5) == 32
print("Basic test do work.")
```

### $\color{#FF66B2}{\textbf{Error tests}}$

```python
# Division by zero
try:
    calculate(1, "/", 0)
    assert False, "CalculatorError should be raised for division by zero"
except CalculatorError as e:
    assert "zero" in str(e).lower()

# Not allowed operator
try:
    calculate(1, "???", 2)
    assert False, "CalculatorError should be raised for an unknown operator"
except CalculatorError as e:
    assert "unknown" in str(e).lower()

# Type error (e.g., text instead of a number)
try:
    calculate("a", "+", 2)
    assert False, "TypeError should be raised for non-numeric arguments"
except TypeError:
    pass

print("Error handling works.")
```
